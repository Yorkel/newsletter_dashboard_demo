# MERGE 

```python
cols = ["url", "title", "date", "text", "from", "type"]

epi = epi[cols]
fed = fed[cols]
fft = fft[cols]
gov = gov[cols]
nuffield = nuffield[cols]
schoolsweek = schoolsweek[cols]
```

```python
full_retro = pd.concat(
    [epi, fed, fft, gov, nuffield, schoolsweek],
    ignore_index=True
)
```

```python
print("Total rows:", len(full_retro))
print(full_retro["from"].value_counts())
print(full_retro["type"].value_counts())
```

```python
#ensure date is datetime 
full_retro["date"] = pd.to_datetime(full_retro["date"], errors="coerce")
```
