```python
import pandas as pd
```

```python
# Load your merged dataset (the one with organisations + links)
df = pd.read_csv("/workspaces/ERP_Newsletter/data/data04_full_articles_scraped/newsletter_full_articles_with_items.csv")
```

```python
# Your shortlist
target_orgs = [
    "schools_week",
    "uk_government",
    "guardian",
    "epi",
    "nfer",
    "conversation",
    "bera",
    "fft_ed_datalab",
    "nuffield",
    "teacher_tapp",
    "bbc",
    "fed",
    "ifg",
    "childrens_commissioner",
    "upen",
    "ippr",
    "chartered_college_of_teaching",
    "5rights_foundation",
    "ukri",
    "ippo",
    "nesta",
    "national_audit_office",
    "cct",
    "national_literacy_trust",
    "british_academy",
    "ifs",
    "uk_parliament",
]
```

```python
# Filter rows for those organisations
df_filtered = df[df["organisation"].isin(target_orgs)].copy()
```

```python
# Check how many rows per organisation
counts = df_filtered["organisation"].value_counts()
print(counts)
```

```python
# Extract unique URLs for each organisation
links_by_org = (
    df_filtered.groupby("organisation")["link_canonical"]
    .unique()
    .reset_index()
)

links_by_org
```

```python
gov_df = df[df["organisation"] == "uk_government"].copy()
```

```python
# Extract unique links
gov_links = gov_df["link_canonical"].dropna().unique()
```

```python
# Save
pd.DataFrame({"uk_government_links": gov_links}).to_csv(
    "/workspaces/ERP_Newsletter/data/data04_full_articles_scraped/uk_government_links.csv",
    index=False
)
```
